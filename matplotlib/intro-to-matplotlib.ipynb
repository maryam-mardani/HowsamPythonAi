## **‚úçÔ∏è Example: Sigmoid**

$$y=\sigma(x)=\frac{1}{1+\exp(-ax)}$$

```python
def sigmoid(x, a):
    y = 1 / (1 + np.exp(-a*x))
    return y
```

```python
x = np.linspace(-10, 10, 1000)
y = sigmoid(x, a=5)

plt.plot(x, y)
```

```python
x = np.linspace(-10, 10, 1000)

plt.plot(x, sigmoid(x, a=0.1))
plt.plot(x, sigmoid(x, a=1))
plt.plot(x, sigmoid(x, a=2))
plt.plot(x, sigmoid(x, a=5))
plt.plot(x, sigmoid(x, a=10))
```

### `scatter3D()`

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 60)
y = np.linspace(-6, 6, 60)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

ax = plt.axes(projection='3d')
ax.scatter3D(X, Y, Z)
```

### `plot_surface()`

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)
```

### 3D plot style

#### ‚Ü≥ Title, label, and etc.

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f')

ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_ylim(-10, 10)

ax.set_title('sine')
```
